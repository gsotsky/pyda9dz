#### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.dat', sep = '::', engine='python', names = ['userId', 'movieId', 'rating', 'timestamp'])
ratings.head()
```

```python
movies = pd.read_csv('movies.dat', sep = '::', engine='python', names = ['movieId', 'movieName', 'moveType'])
movies.head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
five = joined['rating'].isin(['5'])
filtered_joined = joined[five]
```

```python
filtered_joined.head()
```

```python
print('Больше всего поставили пятерок фильму: ')
filtered_joined['movieName'].value_counts().head(1)
```
