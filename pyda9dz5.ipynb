m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку;


```python
def move(directories):
    doc_for_mov = input('Введите номер документа для перемещения ')
    target_dir = input('Введите номер целевой папки ')
    temp = 1

    for docs_nn1 in directories.values():
        for num5 in docs_nn1:
            if num5 == doc_for_mov:
                for dir1 in directories:
                    if dir1 == target_dir:
                        temp = 0
                        temp_list = directories[target_dir]
                        temp_list.append(doc_for_mov)                     
                        directories[target_dir] = temp_list 
                        docs_nn1.remove(doc_for_mov)
                        
                        break

    if temp == 0:
        print('Документ № ', doc_for_mov, 'перемещен в папку ', target_dir)
        print('Новая структура папок: \n', directories)        
    else:
        print("Не правильно введен номер документа или папки")
        print('Cтруктура папок осталась преждней: \n', directories)
```

a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку.

```python
def new_document_add(documents, directories):
    new_doc_add_num = input('Введите номер документа ')
    new_doc_add_type = input('Введите тип документа ')
    new_doc_add_name = input('Введите ФИО владельца документа ')
    new_target_dir = input('Введите номер целевой папки ')
    new_doc_add = {'type': '', 'number': '', 'name': ''}
    new_doc_add['number'] = new_doc_add_num
    new_doc_add['type'] = new_doc_add_type
    new_doc_add['name'] = new_doc_add_name
    temp2=1

    for dir2 in directories:
        if dir2 == new_target_dir:
            temp_list1 = directories[new_target_dir]
            temp_list1.append(new_doc_add_num)
            temp2 = 0

    if temp2 ==0:
        directories[new_target_dir] = temp_list1
        documents.append(new_doc_add)
        print('Документ № ', new_doc_add_num, 'добавлен в папку ', new_target_dir, '. Новая структура папок: \n', directories, '\n Новая структура документов: ', documents)
    else:
        print('Папки с номером ', new_target_dir, 'не существует')
```

```python
def main():
    while True:
        user_input = input('Введите команду: \n \t "p" - узнать человека по номеру документа \n \t "s" - узнать по номеру документа номер полки \n \t "l" -  вывести список всех документов \n \t "as" - добавить новую папку \n \t "d" - удалить документ по номеру \n \t "m" - переместить документ по номеру в целевую папку \n \t "a" - добавить новый документ \n \t "q" - выход из программы \n')
        if user_input == 'p':
            num_docs(documents)
        elif user_input == 's':
            num_dir(directories)
        elif user_input == 'l':
            list_docs(documents)
        elif user_input == 'as':
            new_dir_num(directories)
        elif user_input == 'd':
            del_doc_num(directories, documents)   
        elif user_input == 'm':
            move(directories) 
        elif user_input == 'a':
            new_document_add(documents, directories)    
        elif user_input == 'q':
            print('До свидания')
            break
```
